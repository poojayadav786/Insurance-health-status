## Mutivariate Analysis
Multivariate Data Analysis refers to any statistical technique used to analyze data that arises from more than one variable. This essentially models reality where each situation, product, or decision involves more than a single variable.The techniques are especially valuable when working with correlated variables.

```python
plt.figure(figsize=(10,5))
row=data1.iloc[0,2:]
row1=data1.iloc[3,2:]
row2=data1.iloc[12,2:]
row3=data1.iloc[15,2:]
N=10
ind=np.arange(N)
bar_width=0.3
plt.bar(ind,row,width=bar_width,color='green',zorder=2)
plt.bar(ind+bar_width,row1,width=bar_width,color='yellow',zorder=2)
plt.bar(ind+2*bar_width,row2,width=bar_width,color='black',zorder=2)
plt.bar(ind+3*bar_width,row3,width=bar_width,color='blue',zorder=2)
plt.title("male v/s black non-haspanic v/s Age 18 to 64 yrs v/s income less than 100 percent")
s=(data1.columns[2:])
plt.xticks(ind+bar_width,s,rotation=90)
plt.ylabel(' y axis')
purple_patch=mpatches.Patch(color='green',label='male')
blue_patch=mpatches.Patch(color='yellow',label='Black non-haspanic')
black_patch=mpatches.Patch(color='black',label='Age 18 to 64 yrs')
blue_patch=mpatches.Patch(color='blue',label='Income less than 100 percent')
plt.legend(handles=(purple_patch,blue_patch,black_patch,blue_patch))
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10,5))
row=data1.iloc[1,2:]
row1=data1.iloc[3,2:]
row2=data1.iloc[12,2:]
row3=data1.iloc[15,2:]
N=10
ind=np.arange(N)
bar_width=0.3
plt.bar(ind,row,width=bar_width,color='Red',zorder=3)
plt.bar(ind+bar_width,row1,width=bar_width,color='blue',zorder=2)
plt.bar(ind+2*bar_width,row2,width=bar_width,color='green',zorder=2)
plt.bar(ind+3*bar_width,row3,width=bar_width,color='yellow',zorder=2)
plt.title("female v/s black non-haspanic v/s Age 18 to 64 yrs v/s income less than 100 percent")
s=(data1.columns[2:])
plt.xticks(ind+bar_width,s,rotation=90)
plt.ylabel(' y axis')
purple_patch=mpatches.Patch(color='red',label='female')
blue_patch=mpatches.Patch(color='blue',label='Black non-haspanic')
black_patch=mpatches.Patch(color='green',label='Age 18 to 64 yrs')
yellow_patch=mpatches.Patch(color='yellow',label='Income less than 100 percent')
plt.legend(handles=(purple_patch,blue_patch,black_patch,yellow_patch))
plt.grid(axis='y')
plt.show()
```
